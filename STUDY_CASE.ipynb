```python
sns.relplot(y="TotalSteps", x="Calories",data=df)
plt.title('Total Steps  / Calories')
```

# i see a correlation here between Total Steps and Calories , the more active we are the more calories we burn.


```python
sns.relplot(y="TotalTimeInBed", x="TotalMinutesAsleep",data=df)
plt.title('Total time in Bed / Total Minutes A sleep')
```

# Total Minutes Asleep and Total Time in Bed appear to have a linear connection So, if Bellabeat users want to have better sleep, we should think about utilizing notification to go to bed.
